**Exploratory Data Analysis**

    1) Univariate Analysis - When we are analysing only single variable, for example What's the age distribution i.e. Numerical feature, How many males and females are there? i.e. Categorical feature. 
    
    2) Bivariate Analysis - When we are analysing two variables, for example What's the relationship between Salary and Experience (Two numerical features), Average salary by Gender (One numerical and One categorical features) 
    
    3) Multivariate Analysis - When we are analysing three or more than three variables, for example Whats the relationship between Experience and Salary by Gender (Two numerical and one categorical features), What's the sales by Category and Segment (Two categoricals and one numerical)

# **Matplotlib**

**We cannot use Dataframe directly, we can use Series i.e. slice of a column, Lists, Tuples, Dictionaries or Sets**

```python
plt.figure(figsize=(5,1))
for group,data in dataset.groupby('Gender'):
    plt.barh(group,data['ApplicantIncome'].mean())
```

```python
dataset['Property_Area'].unique()
```

```python
property_dict = dict()
for i in set(dataset['Property_Area']):
    property_dict[i] = dataset[dataset['Property_Area']==i].Loan_ID.count()
```

```python
plt.figure(figsize=(5,1))
plt.bar(property_dict.keys(), property_dict.values())
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.title('Property Area vs Count')
plt.show()
```

```python
plt.figure(figsize=(5,1))
plt.pie(property_dict.values(),labels=property_dict.keys())
plt.title('Property Area vs Count')
plt.show()
```
